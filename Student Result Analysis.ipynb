```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("Expanded_data_with_more_features.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# removing/ deleting unnamed column from the set
df= df.drop("Unnamed: 0", axis=1)
df.head(5)
```

Gender Distribution
-----------

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
# From the above table we have analysed that: 
# The number of females in the data is more than the number of males
```

Parent Education
---

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb, annot= True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.figure(figsize=(4,4))

plt.show()
```

```python
# From the above chart we can conclude that the educatio of the parents have a good impact on thier scores
```

Parent Marital Status
--------------

```python
gb1= df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1, annot= True)
plt.title("Relationship between Parent's Marital Status and Student Score")
plt.figure(figsize=(4,4))
plt.show()
```

```python
# From the above chart we have analysed that the marital status of the parents have no impact or negligible impact on the students score
# due to their parents marital status
```

Ethnic Group
--

```python
print(df["EthnicGroup"].unique())
```

```python
groupA= df.loc[(df['EthnicGroup']== "group A")].count()
groupB= df.loc[(df['EthnicGroup']== "group B")].count()
groupC= df.loc[(df['EthnicGroup']== "group C")].count()
groupD= df.loc[(df['EthnicGroup']== "group D")].count()
groupE= df.loc[(df['EthnicGroup']== "group E")].count()


l= ["group A","group B","group C","group D","group E"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

plt.pie(mlist, labels = l, autopct = "%1.2f%%" )
plt.title("Distribution of Ethnic Groups")
plt.show()
```

```python
ax = sns.countplot( data= df, x= 'EthnicGroup')
ax.bar_label( ax.containers[0])
```

```python
# From the above chart we can see that ethnic group C is more as compared to other ethnic groups
```

Sports
--

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="PracticeSport")
ax.bar_label(ax.containers[0])
plt.title("Sports")
plt.show()
```

```python
# From this chart we can see that students who practice sports sometimes are more as compared to others
```

```python
ax = df.groupby(['PracticeSport'], as_index=False)['WritingScore'].sum().sort_values(by='WritingScore',ascending=False)
plt.title("Writing Score")

sns.barplot(x='PracticeSport', y= 'WritingScore', data = ax)
```

```python
ax = df.groupby(['PracticeSport'], as_index=False)['MathScore'].sum().sort_values(by='MathScore',ascending=False)
plt.title("Math Score")
sns.barplot(x='PracticeSport', y= 'MathScore', data = ax)
```

```python
ax = df.groupby(['PracticeSport'], as_index=False)['ReadingScore'].sum().sort_values(by='ReadingScore',ascending=False)
plt.title("Reading Score")
sns.barplot(x='PracticeSport', y= 'ReadingScore', data = ax)
```

```python
# From the above grahps we can see that students who practice sports sometime have scored more in the subjects as compared to other students who practice regularly or never
```
